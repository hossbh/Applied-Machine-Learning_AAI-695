```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.metrics import mean_squared_error
```

```python
# Simulated data

plt.style.use('seaborn-whitegrid')
noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
y = 2 * x + 11 * x**2 + 3 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

### 1) Please plot the noisy data and the polynomial you found (in the same figure). Please use polynomial order of m = 1, 2, 3, 4, 5, 6, 7, 8, respectively.

```python
# m = 1

x = x.reshape(1,len(x))[0,:]
y = y.reshape(1,len(y))[0,:]
coeff = np.polyfit(x,y,1)
xn = np.linspace(-15,15,100)
yn = np.poly1d(coeff)
plt.plot( xn,yn(xn),x,y,'ro')
```

### 2. Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m.

```python
mse_values = []

for i in range(1, 9):
    p = np.polyfit(x.flatten(), y.flatten(), i)  
    y_p = np.polyval(p, x.flatten())
    mse = mean_squared_error(y.flatten(), y_p)
    mse_values.append(mse)
    print('m =', i, 'MSE =', mse)

best_m = np.argmin(mse_values) +1
print('The best m value is:', best_m)
```

```python
# Plot the MSE values
plt.plot(range(1, 9), mse_values, marker='o', color='red')
plt.xlabel('m')
plt.ylabel('MSE')
plt.title('MSE vs. m')
plt.show()
```

The best value for m is 8 due to having the lowest MSE (10104.21) compared to other m values.
